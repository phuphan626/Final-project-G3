```python
# Import Dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from sklearn.linear_model import LinearRegression
```

```python
# Read in the file
file_path=pd.read_csv('consolidated_coin_data.csv')
```

```python
# Convert to DataFrame
coin_df=pd.DataFrame(file_path)
```

```python
# Display the DataFrame
coin_df
```

```python
# Convert the data from last 6 columns
column_list=['Open','High','Low','Close','Volume','Market Cap']
for x in column_list:
    coin_df[x] = coin_df[x].str.replace(',', '').astype(float)
```

```python
# Double check the type for the DataFrame
coin_df.dtypes
```

```python
# Add an Average-medium column for the DataFrame
coin_df['Average']=(coin_df['Open']+coin_df['High']+coin_df['Low']+coin_df['Close'])/4
```

```python
# Display the newly updated DataFrame
coin_df
```

```python
# Set the index to Currency column 
coin_df=coin_df.set_index('Currency')
```

```python
# Filter out for just bitcoin
btc_df=coin_df.loc['bitcoin']
```

```python
# Display the btc_df
btc_df
```

```python
# Select the X and y for the machine
X=coin_df.drop([],axis=1)
y=coin_df['Average']
```

```python
print(X.shape,y.shape)
```

```python
# Create a Linear Regression model
model = LinearRegression()
```

```python
# Fit in the model
model.fit(X,y)
```

```python
print(model)
```

```python
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_) 
```
